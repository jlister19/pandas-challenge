# Notable Trends
* According to the District Summary, more money is budgeted per student at District schools than at charter schools, yet based on the Percentage of Overall Passing students, District schools tend to underperform relative to charter schools.
* Charter schools tend to have fewer students, lower per student spending and higher average test scores.
* Based on the Percentage of Overall Passing students, larger schools (those with more students) tend to underperform relative to smaller schools.

```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas DataFrames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset.  
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
```

```python
# Begin District Summary work here
# school_data_complete
```

```python
NumSchools = school_data["school_name"].nunique()
# NumSchools
```

```python
NumStudents = student_data["student_name"].count()
# NumStudents
```

```python
DistrictBudget = school_data["budget"].sum()
# DistrictBudget
```

```python
AvgMathScore = student_data["math_score"].mean()
# AvgMathScore
```

```python
AvgReadingScore = student_data["reading_score"].mean()
# AvgReadingScore
```

```python
# create the PctPassingMath variable
# PassingMath = student_data.loc[student_data['math_score'] >= 70, ['math_score',]]
PassingMathdf = student_data.loc[student_data['math_score'] >= 70]
# PassingMathdf
CountPassingMath = PassingMathdf["Student ID"].count()
# CountPassingMath
PctPassingMath = CountPassingMath/NumStudents*100
# PctPassingMath
```

```python
# create the PctPassingReading variable
PassingReadingdf = student_data.loc[student_data['reading_score'] >= 70]
# PassingReadingdf
CountPassingReading = PassingReadingdf["Student ID"].count()
# CountPassingReading
PctPassingReading = CountPassingReading/NumStudents*100
# PctPassingReading
```

```python
# Combine the passing dataframes into a single dataset.  
PassingOveralldf = pd.merge(PassingMathdf, PassingReadingdf, on=["Student ID"])
# PassingOveralldf
CountPassingOverall = PassingOveralldf["Student ID"].count()
# CountPassingOverall
PctPassingOverall = CountPassingOverall/NumStudents*100
# PctPassingOverall
```

```python
# School Summary Section Begins Here
# APPROACH
# From the original schools df, create a new df from the schools df, and add a new per student column. store this new df
New_School_Datadf = school_data
# New_School_Datadf
BudgPerStu = New_School_Datadf["budget"]/New_School_Datadf["size"]
# BudgPerStu
New_School_Datadf["Per Student Budget"] = BudgPerStu
# New_School_Datadf
```

```python
# New_Student_Datadf
Passing_Math_BySchooldf = New_Student_Datadf.loc[New_Student_Datadf["math_score"] >= 70, [
    "school_name", "math_score", "reading_score"]]
# Passing_Math_BySchooldf
```

```python
#Passing_Bothdf = New_Student_Datadf.loc[[New_Student_Datadf["math_score"] >= 70 | [New_Student_Datadf["reading_score"] >= 70]], [
    # "school_name", "math_score"]]
Passing_ReadingBySchooldf = New_Student_Datadf.loc[New_Student_Datadf["reading_score"] >= 70, [
    "school_name", "math_score","reading_score"]]
# Passing_ReadingBySchooldf
```

```python
Passing_Bothdf = pd.merge(Passing_Math_BySchooldf, Passing_ReadingBySchooldf,how='inner',left_index=True, right_index=True)
# Passing_Bothdf
Passing_Bothdf = Passing_Bothdf.drop(['school_name_y', 'math_score_y', 'reading_score_y'], axis=1)
Passing_Bothdf = Passing_Bothdf.rename(columns = {'school_name_x':'school_name',"math_score_x": "math_score","reading_score_x": "reading_score"})
# Passing_Bothdf
Passing_Both_Groupeddf = Passing_Bothdf.groupby('school_name')
# Passing_Both_Groupeddf
Passing_Both_Groupeddf = Passing_Both_Groupeddf.count()
# Passing_Both_Groupeddf
Passing_Both_Groupeddf = Passing_Both_Groupeddf.drop(['reading_score'], axis=1)
# Passing_Both_Groupeddf
Passing_Both_Groupeddf = Passing_Both_Groupeddf.rename(columns = {"math_score": "PassedBothCount"})
# Passing_Both_Groupeddf
# merge this into the partial one
```

```python
# merge the above two new dfs into a new df.
#partialschoolsummarydf = pd.merge(New_School_Datadf, New_Student_Data_Meansdf, how="left", on=["school_name", "school_name"])
partialschoolsummarydf = pd.merge(New_School_Datadf, Passing_Readersdf, how="left", on=["school_name", "school_name"])
partialschoolsummarydf = pd.merge(partialschoolsummarydf, Passing_Mathdf, how="left", on=["school_name", "school_name"])
partialschoolsummarydf = pd.merge(partialschoolsummarydf, New_Student_Data_TotalCountsdf, how="left", on=["school_name", "school_name"])
partialschoolsummarydf = pd.merge(partialschoolsummarydf, Passing_Both_Groupeddf, how="left", on=["school_name", "school_name"])
# partialschoolsummarydf

PctPassingReading = partialschoolsummarydf["reading_score_x"]/partialschoolsummarydf["reading_score_y"]*100
# PctPassingReading
# partialschoolsummarydf
PctPassingMath = partialschoolsummarydf["math_score_x"]/partialschoolsummarydf["math_score_y"]*100
# PctPassingMath
PctPassingBoth = partialschoolsummarydf["PassedBothCount"]/partialschoolsummarydf["math_score_y"]*100

partialschoolsummarydf["% Passing Math"] = PctPassingMath
partialschoolsummarydf["% Passing Reading"] = PctPassingReading
partialschoolsummarydf["% Overall Passing"] = PctPassingBoth

# partialschoolsummarydf
```
